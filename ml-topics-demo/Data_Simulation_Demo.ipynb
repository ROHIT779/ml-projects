## Data simulation for different machine learning problems - Regression, Classification, Time Series, and Clustering  
#### Used Numpy, Sklearn, and Faker libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_blobs
from faker import Faker
```

### Simulation for Simple Linear Regression

```python
# create_linear_regression_data - arguments: N = data size, w = function coefficients
def create_linear_regression_data(N, w):
  X = np.random.rand(N, 1) * 10
  y = w[0] + w[1] * X + np.random.randn(N, 1) # y = w0 + w1*X + e
  return (X, y)

# y = 2 + 3X + e
X, y = create_linear_regression_data(100, (2, 3))

plt.figure(figsize = (8, 6))
plt.scatter(X, y, color = 'black')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simulated Regression Data: y = 2 + 3X + e')
plt.show()
```

### Simulation for Time Series

```python
# create_time_series_data - arguments: N = data size, w = trend and seasonal components' weights
def create_time_series_data(N, w):
  time = np.arange(0, N)
  trend = time * w[0]
  seasonal = np.sin(time * w[1])
  noise = np.random.randn(N)
  y = trend + seasonal + noise
  return (time, y)

time, y = create_time_series_data(100, (0.25, 0.2))

plt.figure(figsize = (8, 6))
plt.plot(time, y, color = 'r')
plt.title("Simulated Time Series Data")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

### Simulation for Classification

```python
# Simulated data created for classification - the dataset contains 100 rows, 4 features and 2 response classes
X, y = make_classification(n_samples = 100, n_features = 4, n_classes = 2)
df = pd.DataFrame(X, columns = ["feature1", "feature2", "feature3", "feature4"])
df["target"] = y
df.head()
```

### Simulation for Clustering

```python
X, y = make_blobs(n_samples = 100, n_features = 2, centers = 3, cluster_std = 0.5, random_state = 0)
feature1 = []
feature2 = []
for row in X:
  feature1.append(row[0])
  feature2.append(row[1])

plt.figure(figsize = (8, 6))
plt.scatter(feature1, feature2, c = y)
plt.title("Clusters by Simulated Data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.show()
```

### Fake Data - names, addresses, emails, phone numbers

```python
# create_fake_data - arguments: N = number of data points
def create_fake_data(N):
  faker = Faker()
  names = [faker.name() for _ in range(N)]
  addresses = [faker.address() for _ in range(N)]
  emails = [faker.email() for _ in range(N)]
  phone_numbers = [faker.phone_number() for _ in range(N)]
  df = pd.DataFrame({"Name": names, "Address": addresses, "Email": emails, "Phone Number": phone_numbers})
  return df

data = create_fake_data(5)
data
```
